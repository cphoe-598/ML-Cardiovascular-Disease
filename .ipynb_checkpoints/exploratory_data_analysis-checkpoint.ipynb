```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

# Cardiovascular Disease ML

## Exploratory Data Analysis

---

This dataset was available on Kaggle from user `thedevastator` as `thedevastator/exploring-risk-factors-for-cardiovascular-diseas`.

### Features

| Name            | Description                                          |
|:----------------|:-----------------------------------------------------|
| **age**         | Age of the individual. (int)                         |
| **gender**      | Gender of the individual. (str)                      |
| **height**      | Height of the individual in centimeters. (int)       |
| **weight**      | Weight of the individual in kilograms. (int)         |
| **ap_hi**       | Systolic blood pressure reading. (int)               |
| **ap_lo**       | Diastolic blood pressure reading. (int)              |
| **cholesterol** | Cholesterol level of the individual. (int)           |
| **gluc**        | Glucose level of the individual. (int)               |
| **smoke**       | Smoking status of the individual. (bool)             |
| **alco**        | Alcohol consumption status of the individual. (bool) |
| **active**      | Physical activity level of the individual. (bool)    |

### Targets

| Name            | Description                                           |
|:----------------|:------------------------------------------------------|
| **cardio**      | Presence or absence of cardiovascular disease. (bool) |

# Load Dataset

```python
cv_data = pd.read_csv("./data/heart_data.csv", sep = ",")
cv_data.drop(["index", "id"], axis = 1, inplace = True)

print(f"Rows: {cv_data.shape[0]}, Cols: {cv_data.shape[1]}")
cv_data.head(3)
```

# Quick Cleaning

```python
# verify there are no NULL values
print(f"\nNULL Values Per Column:\n{cv_data.isna().sum()}")

# convert age from days to years
cv_data["age"] = (cv_data["age"] / 365)

# remove unlikely BP values
cv_data.loc[(cv_data["ap_hi"] < 80) | (cv_data["ap_hi"] > 180), "ap_hi"] = cv_data["ap_hi"].median()
cv_data.loc[(cv_data["ap_lo"] < 60) | (cv_data["ap_lo"] > 120), "ap_lo"] = cv_data["ap_lo"].median()
```

# Data Spread

```python
numeric_cols = {
    "age" : "Age (years)",
    "height" : "Height (cm)",
    "weight" : "Weight (kg)",
    "ap_hi" : "Systolic BP (mm HG)",
    "ap_lo" : "Diastolic BP (mm HG)"
}

n_vars = len(numeric_cols)
fig, axs = plt.subplots(n_vars, 1,
                        figsize = (6, 2 * n_vars),
                        constrained_layout = True)

for i, col in enumerate(numeric_cols):
    axs[i].boxplot(cv_data[col],
                   orientation = "horizontal",
                   patch_artist = True,
                   boxprops = dict(facecolor = "lightgrey", color = "black"),
                   medianprops = dict(color = "red"),
                   whiskerprops = dict(color = "black"),
                   capprops = dict(color = "black"),
                   flierprops = dict(marker = ".", markersize = 3, linestyle = "none"))
    axs[i].set_title(numeric_cols[col])
    axs[i].set_yticks([])

plt.show()
```

```python
features, target = cv_data.drop("cardio", axis = 1), cv_data["cardio"]
```

# Linear Regression Analysis

```python
features_intercept = sm.add_constant(features)

# lin_reg = smf.ols(formula = "cardio ~ gender + height + weight + ap_hi + ap_lo + cholesterol + gluc + smoke + alco + active + C(age_group)", data = cv_data).fit()
lin_reg = sm.OLS(target, features_intercept).fit()
lin_reg.summary()
```

# Correlation Matrix

```python
corr_data = cv_data.corr().to_numpy().round(2)

# initialize heatmap
fig, ax = plt.subplots()
corr_heatmap = ax.imshow(corr_data, cmap = "bwr", vmin = -1, vmax = 1)

for i in range(len(cv_data.columns)):
    for j in range(len(cv_data.columns)):
        text = ax.text(j, i, corr_data[i, j],
                       ha = "center", va = "center", size = 6.5)

# labels
ax.set_title("Fig. 1: Correlation Between All Variables, Features and Target")
ax.set_xticks(range(len(cv_data.columns)), labels = list(cv_data.columns),
              rotation = 45, ha = "right", rotation_mode = "anchor")
ax.set_yticks(range(len(cv_data.columns)), labels = list(cv_data.columns))

plt.colorbar(corr_heatmap)  # legend
plt.show()
```

# Frequencies Of Target Values

```python
target_freqs = target.value_counts()

# initialize
fig, ax = plt.subplots()
target_bars = ax.bar(target_freqs.index, target_freqs,
                     color = ["tab:blue", "tab:red"])

# labels
ax.set_title("Fig. 2: Frequencies Of Healthy And Disease Patients")
ax.set_xticks([0, 1], ["Healthy", "Disease"])
ax.set_ylabel("Frequency")
ax.bar_label(target_bars)

plt.show()
```

# Principal Component Analysis

```python
# initialize PCA
f_model = PCA(n_components = 2)

# compute cumulative explained variance for each PC
f_model.fit_transform(features)
cum_expl_variance = f_model.explained_variance_ratio_.cumsum().round(3) * 100

# reduce features
f_reduced = f_model.fit_transform(features)

# initialize plot
fig = plt.figure(1, figsize = (8, 6))
ax = fig.add_subplot()
features_PCA = ax.scatter(
    f_reduced[:, 0],
    f_reduced[:, 1],
    c = target,
    s = 40,
)

ax.set(
    title = "Fig. 3: PCA - Clustered Observations",
    xlabel = f"PC-1 ({cum_expl_variance[0]}%)",
    ylabel = f"PC-2 ({cum_expl_variance[1]}%)",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])

# Add a legend
legend1 = ax.legend(
    features_PCA.legend_elements()[0],
    list(target.unique()),
    loc = "upper left",
    title = "Cardiovascular Disease?",
)
ax.add_artist(legend1)

plt.show()
```

# Conclusions

According to **Fig. 1**, cardiovascular disease may be positively associated with age, weight, and cholesterol (blood pressure too, but to a lesser degree); and negatively (inversely) associated with physical activity. These will likely be the best predictors of cardiovascular disease. **Fig. 2** shows that there is a balanced distribution of healthy and diseased patients--beneficial for creating a classifier model. Following a dimensionality reduction using PCA (**Fig. 3**), I found that the first two PCs explain (cumulative) 79% of the variance among features.
